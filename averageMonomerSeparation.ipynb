```python
import re
import numpy as np

lammpsOutputDir = '/Users/johnwhitfield/Desktop/localSims/n100Testing/3_1_n100/KNOT_data/XYZ_knot3_1.n100.3200000.xyz'

with open(lammpsOutputDir, 'r') as f:
    lines = f.readlines()

# Skip headers and empty lines
lines = [line for line in lines if line.strip() and not line.strip().isdigit()]

# Store all positions
positions = []
for line in lines:
    cleaned_line = re.sub(r'\s+', ' ', line.strip())
    parts = cleaned_line.split(' ')
    
    # Skip if not enough parts (should be: element x y z)
    if len(parts) < 4:
        continue
        
    # Extract positions - parts[1], parts[2], parts[3] are x,y,z
    try:
        x = float(parts[1])
        y = float(parts[2])
        z = float(parts[3])
        positions.append([x, y, z])
    except (ValueError, IndexError):
        continue

# Calculate distances between consecutive monomers
distances = []
for i in range(1, len(positions)):
    # Current monomer position
    curr_pos = positions[i]
    # Previous monomer position  
    prev_pos = positions[i-1]
    
    # Calculate Euclidean distance
    distance = np.linalg.norm(np.array(curr_pos) - np.array(prev_pos))
    distances.append(distance)

# Calculate average separation
if distances:
    avg_separation = np.mean(distances)
    std_separation = np.std(distances)
    
    print(f"Average monomer separation: {avg_separation:.4f}")
    print(f"Standard deviation: {std_separation:.4f}")
    print(f"Number of bonds measured: {len(distances)}")
    print(f"Min distance: {min(distances):.4f}")
    print(f"Max distance: {max(distances):.4f}")
else:
    print("No distances calculated - check your data format")
```
